```python
import plotly.plotly as py
import plotly.graph_objs as go
import pandas as pd 
import inflect
import datetime

p = inflect.engine() # Turn 1 into 1st.. and so on

# Getting Data - this table should be updated before-hand
tables = pd.read_csv('tables.csv',header=0)

# Names to be displayed on plot - messy, I know.
nameMap={'Arsenal FC':'Arsenal',
 'AFC Bournemouth':'Bournemouth',
 'Brighton & Hove Albion':'Brighton',
 'Burnley FC':'Burnley',
 'Chelsea FC':'Chelsea',
 'Crystal Palace FC':'Crystal Palace',
 'Everton FC':'Everton',
 'Huddersfield Town':'Huddersfield',
 'Leicester City FC':'Leicester',
 'Liverpool FC': 'Liverpool',
 'Manchester City FC':'Man City',
 'Manchester United FC':'Man United',
 'Newcastle United FC':'Newcastle',
 'Southampton FC':'Southampton',
 'Stoke City FC':'Stoke',
 'Swansea City FC':'Swansea',
 'Tottenham Hotspur FC':'Spurs',
 'Watford FC': 'Watford',
 'West Ham United FC':'West Ham',
 'West Bromwich Albion FC':'West Brom'}

# Getting most currennt matchday
currMatchday = tables['matchday'].max() # Don't scrape matchdays which haven't yet happened
currTable = tables[tables['matchday'] == currMatchday] 

# Maybe I could use map instead..
currTable['div'] = '1-4'
currTable.loc[currTable['position'] > 4,'div'] = '5-11'
currTable.loc[currTable['position'] > 11,'div'] = '12-17'
currTable.loc[currTable['position'] > 17,'div'] = '18-20'

# Mapping colors
colors = dict(zip(currTable['div'].unique(),['green','yellow','orange','red']))

# Contains data for plotly.iplot
traces = []

# Point Growth - I wanted to graph growth of points over time
# Decided to not pursue this
'''
for team, teamTable in tables.groupby('teamName'):
    traces.append(
        go.Scatter( 
        x = teamTable['matchday'],
        y = teamTable['goals'],
        mode = 'lines+markers',
        name = team,
            line = dict(
                color = colors[team]
        
            )
        )
    )
'''

# Goals for & Against 
# Plotting by division... so it can be sorted so
for div, divTable in currTable.groupby('div',sort=False):
    traces.append(
        go.Scatter(
            x = divTable['goalsAgainst'],
            y = divTable['goals'],
            mode = 'markers',
            name=div,
            hovertext= divTable.position.apply(p.ordinal),
            hoverinfo = 'x+y+text',
            textposition = 'bottom',
                marker = dict(
                    color = colors[div],
                    size = 12,
                    line = dict(
                           width=2
                    )
                )

        )

    )

# Annotation Data
ann_data = [tuple(row) for row in currTable[['teamName','goals','goalsAgainst']].values]
layout = go.Layout(title = 'Goals For/Against 17-18 <br> {} :: @FootyNumb3rs'.format(datetime.date.today()),
                   hovermode='closest',
                    titlefont=dict(
                    size=18,
                    color='#7f7f7f'),
              xaxis = dict(title = 'Goals Against',
                           titlefont=dict(size=18),
                          range=[45,5]),
              yaxis = dict(title = 'Goals For',
                           titlefont=dict(size=18),
                           range=[5,60]),
             annotations = [dict(
                 xref='x',
                 yref='y',
                 x = ga,
                 y = gf-1.25,
                 text = nameMap[name],
                 showarrow=False
             )for name, gf,ga in ann_data],
              height = 800,
              width = 900,
              images=[dict(
                      source="https://upload.wikimedia.org/wikipedia/en/f/f2/Premier_League_Logo.svg",
                      xref="paper", yref="paper",
                      x=.18, y=1.038,
                      sizex=0.18, sizey=0.18,
                      xanchor="right", yanchor="bottom"
              )],
              legend = dict(traceorder='normal'))
                   
figure = go.Figure(data=traces,layout=layout)
py.iplot(figure,filename='for-and-against')
```
